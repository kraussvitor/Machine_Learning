## T1) *Kernel* linear em SVM

O objetivo dessa questão é provar que o *kernel* linear $k(\mathbf{x}, \mathbf{y}) = \langle \mathbf{x}, \mathbf{y} \rangle$ é de fato um *kernel*, ou seja, uma função positiva-semidefinida. Para isso, mostre que a sua matriz de Gram associada $\mathbf{G}$, definida por $$\mathbf{G}_{ij} = \langle \mathbf{x}_i, \mathbf{x}_j \rangle$$ é simétrica e positiva-semidefinida, para todo conjunto de vetores $\mathbf{x}_1, \dots, \mathbf{x}_n \in \mathbb{R}^p$.

Conclua que $\mathbf{G}$ é positiva-definida se e somente se $\mathbf{x}_1, \dots, \mathbf{x}_n \in \mathbb{R}^p$ são linearmente independentes. Isso pode acontecer para qualquer relação entre $n$ e $p$?


-- Escreva aqui a sua resolução --

Primeiramente, pela simetria do produto interno, temos que 

$$\mathbf{G}_{ij} = \langle \mathbf{x}_i, \mathbf{x}_j \rangle = \langle \mathbf{x}_j, \mathbf{x}_i \rangle = \mathbf{G}_{ji} $$

 e, portanto, a matriz $\mathbf{G}$ é simétrica. 

Agora, seja $\mathbf{z} = [z_1, z_2, \ldots, z_n]^T \in \mathbb{R}^n$ um vetor qualquer. 

Note que, em notação vetorial/matricial, $\mathbf{G}_{ij} = \langle \mathbf{x}_i, \mathbf{x}_j \rangle = x_i^T x_j$.

Com isso, note que 

$$ \mathbf{z}^T \mathbf{G} \mathbf{z} = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \mathbf{G}_{ij} z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i (x_i^T x_j) z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} (z_i x_i^T) (x_j z_j) = \Big( \sum_{i=1}^{n} z_i x_i^T \Big) \Big( \sum_{j=1}^{n} x_j z_j \Big) = \mathbf{v}^T \mathbf{v} = \langle \mathbf{v}, \mathbf{v} \rangle \geq 0,  \\ \forall \mathbf{v} \in \mathbb{R}^n$$

onde $\mathbf{v} = \sum_{i=1}^{n} x_i z_i$.

Logo $\mathbf{z}^T \mathbf{G} \mathbf{z} \geq 0, \forall \mathbf{z} \in \mathbb{R}^n$ e, portanto, $\mathbf{G}$ é positiva semidefinida para qualquer conjunto $\{ \mathbf{x}_1, \ldots, \mathbf{x}_n \}$. 

Agora suponha que $\mathbf{G}$ é positiva definida, ou seja, que $\mathbf{z}^T \mathbf{G} \mathbf{z} > 0, \forall \mathbf{z} \neq \vec{0} \in \mathbb{R}^n$.

Suponha, para uma contradição, que $\exists \ell, q,  1 \leq \ell, q \leq n, \ell \neq q$ tais que $\langle \mathbf{x}_{\ell}, \mathbf{x}_{q} \rangle = 0 $, ou seja, $\mathbf{x}_{\ell}$ e $\mathbf{x}_{q}$ não são linearmente independentes. 

Considere o vetor $\mathbf{u} \in \mathbb{R}^n$ tal que $u_{\ell} = u_q = 1$ e $u_i = 0$ para $i \neq \ell,q$.

Daí, $$ \mathbf{u}^T \mathbf{G} \mathbf{u} = \sum_{i=1}^{n} \sum_{j=1}^{n} u_i \mathbf{G}_{ij} u_j = \mathbf{G}_{\ell q} + \mathbf{G}_{q \ell} = 2 \langle \mathbf{x}_{\ell}, \mathbf{x}_{q} \rangle = 0  $$

Assim mostramos um vetor $\mathbf{u} \neq \vec{0}$ tal que $\mathbf{u}^T \mathbf{G} \mathbf{u} = 0$, uma contradição.

Portanto $\langle \mathbf{x}_{i}, \mathbf{x}_{j} \rangle \neq 0, \forall i,j, 1 \leq i,j \leq n$, ou seja, os vetores $\mathbf{x}_1, \ldots, \mathbf{x}_n$ são linearmente independentes. 

Finalmente, suponha que os vetores $\mathbf{x}_1, \ldots, \mathbf{x}_n$ são linearmente independentes, ou seja, $\langle \mathbf{x}_{i}, \mathbf{x}_{j} \rangle \neq 0, \forall i,j, 1 \leq i,j \leq n$. 

Como já vimos, para qualquer $\mathbf{z} \in \mathbb{R}^n$, $$\mathbf{z}^T \mathbf{G} \mathbf{z} = \langle \mathbf{v}, \mathbf{v} \rangle$$

onde $\mathbf{v} = \sum_{i=1}^{n} x_i z_i$.

Logo, pelas propriedades do produto interno, $ \mathbf{z}^T \mathbf{G} \mathbf{z} = 0 \iff \mathbf{v} = \vec{0}$.

Como $\mathbf{v} = \sum_{i=1}^{n} x_i z_i$ e os vetores  $\mathbf{x}_1, \ldots, \mathbf{x}_n$ são linearmente independentes, então $\mathbf{v} = \vec{0} \iff \mathbf{z} = \vec{0}$. Logo $\mathbf{z}^T \mathbf{G} \mathbf{z} = 0 \iff \mathbf{z} = \vec{0}$.
Portanto $\mathbf{G}$ é positiva definida(lembre que já havíamos mostrado que ela era positiva semidefinida). 

Observe que quando $n > p$, então necessariamente existem $\mathbf{x}_{\ell}$ e $\mathbf{x}_{q}$ linearmente dependentes e, com isso, a matriz deixa de ser positiva definida e passa a ser apenas semidefinida. Portanto não, não pode acontecer para qualquer relação entre $n$ e $p$. A matriz $\mathbf{G}$ só pode ser positiva definida se $n \leq p$.

## T2) *Kernel* radial em SVM

Agora vamos deixar um pouco mais emocionante! Vamos provar que o *kernel* radial $k(x, y) = \exp(-\gamma |x - y|^2)$ é uma função positiva-semidefinida. Note que, como $x$ e $y$ não estão em negrito, estamos considerando $x, y \in \mathbb{R}$. O caso geral em $\mathbb{R}^p$ é análogo. Para facilitar a escrita, considere $\gamma = 1/2$, porém o resultado vale para $\gamma > 0$ qualquer.

O raciocínio é o mesmo que o da questão anterior: mostre que a matriz de Gram $\mathbf{G}$ associada à $k$ é simétrica e positiva-semidefinida, onde $$\mathbf{G}_{ij} = k(x_i, x_j),$$ para todos $x_1, \dots, x_n \in \mathbb{R}$, porém fazer isso agora é um pouco mais divertido.

Uma dica é lembrar de números complexos e também que a função característica de uma variável aleatória normal padrão é dada por $$\mathbb{E}[e^{itZ}] = \exp(-t^2/2).$$ Juro que dá certo!

-- Escreva aqui a sua resolução --

Primeiramente, pela simetria do valor absoluto, temos que 

$$\mathbf{G}_{ij} = \exp\Bigg\{{\frac{-\left\vert x_i - x_j \right\vert ^2}{2}} \Bigg\} = \exp\Bigg\{{\frac{-\left\vert x_j - x_i \right\vert ^2}{2}} \Bigg\} = \mathbf{G}_{ji}$$

Logo $\mathbf{G}$ é simétrica.

Novamente, seja $\mathbf{z} = [z_1, \ldots, z_n]^T \in \mathbb{R}^n$ um vetor qualquer. Temos que 

$$ \mathbf{z}^T \mathbf{G} \mathbf{z} = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \mathbf{G}_{ij} z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \exp\Bigg\{{\frac{-\left\vert x_i - x_j \right\vert ^2}{2}} \Bigg\} z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \exp\Bigg\{{\frac{-(x_i - x_j) ^2}{2}}\Bigg\} z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \mathbb{E}[e^{i(x_i - x_j)Z}] z_j = \sum_{i=1}^{n} \sum_{j=1}^{n} z_i \mathbb{E}[e^{i(x_i - x_j)Z}] z_j = \mathbb{E} \Big[ \sum_{i=1}^{n} \sum_{j=1}^{n} z_i e^{i(x_i - x_j)Z} z_j \Big] = \mathbb{E} \Big[ \sum_{i=1}^{n} \sum_{j=1}^{n} (z_i e^{i x_i Z}) (e^{-i x_j Z} z_j) \Big] = \mathbb{E} \Big[ \Big( \sum_{i=1}^{n} z_i e^{i x_i Z} \Big) \Big( \sum_{j=1}^{n} e^{-i x_j Z} z_j \Big) \Big] = \mathbb{E} [ w \overline{w} ] \geq 0 , \forall w \in \mathbb{C} $$

onde $w = \sum_{i=1}^{n} z_i e^{i x_i Z} \in \mathbb{C}$.

Logo $\mathbf{z}^T \mathbf{G} \mathbf{z} \geq 0, \forall \mathbf{z} \in \mathbb{R}^n$ e, portanto, $\mathbf{G}$ é positiva semidefinida.

## P1) Usando *kernel* personalizados no sklearn - Parte 01

Explique abaixo como usar o seu próprio *kernel* no sklearn. Note que há duas formas de fazer isso: chamando uma função do Python que representa o *kernel* ou informando a sua matriz de Gram associada. Explique como as duas funcionam, explicitando as diferenças, e argumentando quando uma é preferível em relação à outra.

-- Escreva aqui a sua resolução --

Começamos pelo caso em que o próprio kernel é passado como argumento.

O primeiro passo é definir este kernel como uma função do Python. Esta função deve receber como argumento duas matrizes $n \times p$, onde $n$ é o número de vetores $\mathbf{x}_1, \ldots, \mathbf{x}_n$ e $p$ a dimensão destes vetores. 

Como exemplo vamos considerar o kernel linear $\mathbf{k}(\mathbf{x}, \mathbf{y}) = \langle \mathbf{x}, \mathbf{y} \rangle $.

```python
# Seu código aqui
def custom_kernel(x, y):
  return np.dot(x, y.T)
```

É ideal que a sua função kernel esteja vetorizada, assim como a função acima. Note que não tivemos que iterar pelas posições da matriz. Isto permite computações vetoriais, garantindo uma performance melhor, principalmente em arquiteturas projetadas para este tipo de operação, como GPU's. 

Note que, dados os vetores $\mathbf{x_i}, \ldots, \mathbf{x_n}$, se $X$ é a matriz cujas linhas são estes vetores, então a nossa função kernel vetorizada já é capaz de calcular a matrix de Gram.

Abaixo segue um exemplo:

```python
# Exemplo de evaluação do kernel
NUM_SAMPLES = 3
p = 2
mu = np.zeros((p), float)
cov = np.diag(np.ones((p), float))
X = np.random.multivariate_normal(mu, cov, size=NUM_SAMPLES)
print("X = \n", X)
print("\nMatriz G = \n", custom_kernel(X, X))
```

Uma vez definido o nosso kernel, basta instanciar um objeto da classe *svm.SVC* e passar nosso kernel como argumento para o construtor.

```python
svm_with_custom_kernel = sklearn.svm.SVC(kernel=custom_kernel)
```

Agora vamos tratar do caso em que a matriz de Gram é passada como argumento para o construtor. 

Para isso, nós devemos ter uma variável de tipo matriz, por exemplo um objeto da classe numpy.ndarray ou um objeto de alguma das classes oferecidas pelo scipy.sparse para matrizes esparsas. 

O primeiro passo é declarar e preencher esta matriz.

Primeiro, suponha que temos uma função *my_kernel* não vetorizada. Neste caso, devemos iterar pelas posições da matriz e preencher cada uma indivudualmente.



```python
# Se my_kernel for uma função não vetorizada
def get_gram_matrix(my_kernel, X):
  '''kernel deve ser uma função/callable. X deve ser uma lista de vetores'''
  n = X.shape[0] # num de colunas
  gram_matrix = np.zeros((n, n), float)
  for i in range(n):
    for j in range(n):
      gram_matrix[i][j] = my_kernel(X[i, :], X[j, :])
  return gram_matrix
```

```python
custom_gram_matrix = get_gram_matrix(custom_kernel, X)
print("Matriz G = \n", custom_gram_matrix)
```

Como vimos, a nossa função kernel também pode já estar vetorizada. Daí, basta chamar diretamente a função kernel na matriz $X$ cujas linhas são os vetores $\mathbf{x}_i$.

```python
custom_gram_matrix = custom_kernel(X, X)
print("Matriz G = \n", custom_gram_matrix)
```

Uma vez que já temos a matriz de Gram associada ao nosso kernel, basta passá-la como argumento para o construtor da classe *svm.SVC*.

```python
svm_with_custom_gram = sklearn.svm.SVC(kernel='precomputed')
```

Uma vez que já instanciamos um objeto da classe *svm.SVC*, o seu uso varia um pouco dependendo se passamos como argumento a função kernel ou sua matriz de Gram.

Seja $X$ a matriz já descrita acima e $y$ o vetor coluna da variável resposta associada a cada vetor $\mathbf{x}_i$.

Por exemplo, para o método *fit* da classe, se nós passamos a função kernel, basta passar $X$ e $y$ como argumentos. Se nós passamos a matriz de Gram, ela então deve ser passada como argumento no lugar de $X$ na chamada de *fit*. 

```python
y = np.array([0, 1, 0]).transpose() # Gerar respostas aleatórias
svm_with_custom_kernel.fit(X, y)
svm_with_custom_gram.fit(custom_gram_matrix, y)
```

Similarmente, quando queremos fazer previsões, se nós passamos a função kernel, basta chamar o método *predict* na matrix $P$, que é $m \times p$, na qual as linhas da matriz são os $m$ vetores para os quais queremos fazer previsões. Sejam $\mathbf{p}_1, \ldots, \mathbf{p}_m$ as linhas da matriz $P$. Se a matriz de Gram foi passada, então devemos passar como argumento a matrix $K^P$ tal que $K_{ij}^{P} = k(\mathbf{p}_i, \mathbf{x}_j)$.

No caso do kernel linear, $K^P = PX^T$.

```python
P = np.random.multivariate_normal(mu, cov, size=NUM_SAMPLES)
print("Predições usando custom_kernel: ", svm_with_custom_kernel.predict(P))
print("Predições usando matriz de Gram: ", svm_with_custom_gram.predict(np.dot(P, X.T)))
```

Para mostrar qual a principal diferença entre como argumento a função kernel ou a matriz de Gram, vamos modificar um pouco a nossa função *custom_kernel* para que ela printe uma mensagem toda vez que for chamada. 

```python
def custom_kernel_with_message(x, y):
  print("\t ---- custom kernel foi chamado ---- \n")
  return np.dot(x, y.T)
```

```python
svm_with_message = svm.SVC(kernel=custom_kernel_with_message)
```

```python
svm_with_message.fit(X, y)
```

```python
svm_with_message.predict(P)
```

Note que quando instanciamos um objeto tendo passado a função kernel, o construtor não fez nenhuma chamada a função kernel, ou seja, ainda não instanciou a matriz de Gram. Observe também que quando chamamos métodos como *fit* e *predict*, que precisam da matriz de Gram, estes métodos chamam a nossa função kernel. 

Podemos ver na linha 86 do código fonte da classe *BaseLibSVM*(disponível em https://github.com/scikit-learn/scikit-learn/blob/master/sklearn/svm/_base.py), que é a classe base para as demais classes do *sklearn.svm*, que o argumento *kernel* passado para o construtor é armazenado em uma variável membro chamada *kernel*, que é do tipo *kernel*.

Consultando o código fonte da classe *kernel*(disponível em https://github.com/scikit-learn/scikit-learn/blob/master/sklearn/svm/src/libsvm/svm.cpp), podemos ver nas linhas 354 a 371 que esta classe armazena apenas uma referência para a função kernel ou matriz de Gram que foi passada como argumento para o construtor. 

No caso em que passamos a função kernel como argumento, esta referência é então para uma função. No caso em que passamos a matriz de Gram como argumento, a referência é para uma matriz. 

Também podemos ver na linha 239 e na linha 320 do código fonte da classe *BaseLibSVM* que os métodos membros *fit* e *predict* chamam o método membro *_compute_kernel*.

Nas linhas 364 a 373 podemos ver que o método *_compute_kernel* verifica se o que é referenciado pela variável membro *kernel* é "callable" ou não, ou seja, se é uma função ou não. Se é, de fato, uma função, então ele chama esta função na matriz $X$ passada como argumento. Se não é uma função, então ele simplesmente retorna a matriz passada como argumento. Isso explica por que, como exemplificado acima, em um caso nós devemos passar a matriz $X$ e no outro nós devemos passar a matriz de Gram. 

Concluímos então que se nós passamos uma função kernel(chame esse de primeiro caso) então toda vez que for necessária a matriz de Gram associado ao modelo ela será calculada na hora, chamando função kernel passada. Por outro lado, se o que foi passado foi a própria matriz de Gram(chame esse de segundo caso), então ela não precisa ser novamente calculada e basta ser acessada na memória. 

Quais são, portanto, as vantagens e desvantagens de cada método?

O primeiro caso é mais adequado para situações dinâmicas, nas quais o conjunto de dados/matriz de Gram varia. Por exemplo, se quisermos fazer $k$-fold cross validation, teremos que, para cada uma das $k$ iterações,  calcular respectiva a matriz de Gram. 

O segundo caso é mais adequado para situações estáticas, nas quais o conjunto de dados/matriz de Gram não muda, e em situações nas quais o cálculo da matriz de Gram demanda mais tempo, por exemplo quando o conjunto de dados é muito grande. Nestas situações, é mais eficiente que a matriz de Gram seja calculada apenas uma vez e, quando necessário(por exemplo quando chamamos *fit* ou *predict*), basta acessá-la na memória.  

## P2) Usando *kernel* personalizados no sklearn - Parte 02

Relembremos o exemplo feito em sala de aula com a função 'make_moons' do sklearn. Gere um conjunto de dados análogo àquele e tente classificá-lo usando SVM com o *kernel de Ornstein-Uhlenbeck*, dado por $$k(\mathbf{x}, \mathbf{y}) = \exp(-\|\mathbf{x} - \mathbf{y}\|/\ell).$$

Compare os resultados obtidos com esse *kernel* e com o *kernel* radial; em particular, plote e compare as fronteiras de decisão de ambos os classificadores. Além disso, interprete esse *kernel*, bem como o seu parâmetro $\ell$, que você é livre para escolher como quiser, desde que arque com as possíveis consequências de sua escolha :-)

Comecemos com a interpretação do kernel. 

Primeiro, o vetor diferença $\mathbf{x} - \mathbf{y}$ mede a diferença em cada componente entre $\mathbf{x}$ e $\mathbf{y}$. Quão mais "diferentes" forem os vetores $\mathbf{x}$ e $\mathbf{y}$, maior serão os valores absolutos de cada componente de $\mathbf{x} - \mathbf{y}$. Ou seja, $\mathbf{x} - \mathbf{y}$ é uma medida vetorial de quão diferentes são os vetores $\mathbf{x}$ e $\mathbf{y}$. 

Além disso, sabemos que o operador norma $\| \circ \|$ mede o "tamanho" de um vetor. Pela definição da norma(euclidiana) $\| \mathbf{v} \| = \sqrt{\sum_{i=1}^{p} v_i^2}$, note que quão maiores os valores absolutos das componentes $v_i$ do vetor, maior a sua norma. Como vimos, quão mais diferentes são $\mathbf{x}$ e $\mathbf{y}$, maior o valor absoluto de cada componente de $\mathbf{x} - \mathbf{y}$ e, consequentemente, maior o valor de $\| \mathbf{x} - \mathbf{y} \|$. Assim o operador norma transforma nossa medida vetorial $\mathbf{x} - \mathbf{y}$ da diferença entre $\mathbf{x}$ e $\mathbf{y}$ em uma medida escalar $\| \mathbf{x} - \mathbf{y} \|$ desta diferença. 

Note que $\| \mathbf{x} - \mathbf{x} \| = 0$, ou seja, sob esta interpretação, a diferença entre um vetor e ele mesmo é zero, a menor possível(pois a norma é sempre maior ou igual a zero). 

Assim, dado $\ell > 0$, note que $k(\mathbf{x}, \mathbf{y})$ é decrescente em $\| \mathbf{x} - \mathbf{y} \|$, isto é, quanto maior a diferença entre $\mathbf{x}$ e $\mathbf{y}$, menor o valor de $k(\mathbf{x}, \mathbf{y})$. Portanto $k(\mathbf{x}, \mathbf{y})$ é uma medida escalar da similaridade entre os vetores $\mathbf{x}$ e $\mathbf{y}$. Observe que $k(\mathbf{x}, \mathbf{x}) = 1$, o maior valor possível para este kernel. Ou seja, sob esta interpretação, o nível de similaridade entre um vetor e ele mesmo é o maior possível. 

Agora qual a função do parâmetro $\ell$?

Suponha $\ell = 1$.

Considere um conjunto $X \subset \mathbb{R}^p$ com pelo menos dois pontos. 

Sejam $\mathbf{x}, \mathbf{y} \in X$. Temos então que $k(\mathbf{x}, \mathbf{y}) = \exp(-\|\mathbf{x} - \mathbf{y}\|)$.

Agora consideremos o conjunto $X^{\prime} = \{ 2x \mid x \in X \}$, ou seja, o conjunto dos pontos de $X$ multiplicados por dois. As estruturas e as topologias de $X$ e $X^{\prime}$ são similares. 

Sejam $\mathbf{x}^{\prime}, \mathbf{y}^{\prime} \in X$ tais que $\mathbf{x}^{\prime} = 2\mathbf{x}, \mathbf{y}^{\prime} = 2\mathbf{y}$. Temos que $\| \mathbf{x}^{\prime} - \mathbf{y}^{\prime} \| = 2\| \mathbf{x} - \mathbf{y} \|$. 

Assim, temos que 

$$k(\mathbf{x}^{\prime}, \mathbf{y}^{\prime}) = \exp(-2\|\mathbf{x} - \mathbf{y}\|) \leq k(\mathbf{x}, \mathbf{y}) = \exp(-\|\mathbf{x} - \mathbf{y}\|)$$

Assim, uma simples mudança na escala dos dados causou uma mudança na nossa medida de similaridade. Isto seria algo como mudar alguma noção de distância quando convertéssemos metros para quilômetros, o que é estranho e inadequado. Portanto o primeiro motivo para inclusã de um parâmetro $\ell$ é normalização. No exemplo anterior, poderiamos considerar um parâmetro $\ell_1$ para $X$ e $2\ell_1$ para $X^{\prime}$. Com isso, obteríamos kernels equivalentes. 

Lembremos a função de densidade de uma distribuição exponencial $f(x) = \lambda \exp \{ -\lambda x \}$. Sabemos que a variância desta distribuição é $\frac{1}{\lambda^2}$. Assim, quanto menor o parâmetro $\lambda$, maior a variância da distribuição, ou seja, a sua densidade está mais distribuída no espaço. 

O parâmetro $\ell$ têm efeito similar no kernel. Considere um vetor $\mathbf{x}^{\prime}$ usado para treino. Seja $\lambda = \frac{1}{\ell}$. Assim, vemos que um valor maior de $\ell$ faz com que o kernel $k(\mathbf{x}, \mathbf{x}^{\prime}) = \exp\{ -\lambda \| \mathbf{x} - \mathbf{x} \| \}$ seja mais distribuído pelo espaço, ou seja, a influência do vetor de treino $\mathbf{x}^{\prime}$ no espaço seja maior(note que $k$ é centrado em $\mathbf{x}^{\prime}$). Por outro lado, valores pequenos de $\ell$ fazem com $k(\mathbf{x}, \mathbf{x}^{\prime})$ esteja mais concentrado ao redor de $\mathbf{x}^{\prime}$ e menos distribuído no espaço, ou seja, a influência do vetor de treino $\mathbf{x}^{\prime}$ no espaço é menor. 

Concluímos que o parâmetro $\ell$ ajusta a àrea de influência dos vetores suporte. 

O parâmetro $\ell$ tem portanto efeito no bias-variance tradeoff do modelo e sua escolha deve ser feita com este tradeoff em mente.

```python
# Seu código aqui
def ornstein_uhlenbeck_gram(X, Y, ell):
  n = X.shape[0]
  mat = np.zeros((n, n), float)
  for i in range(n):
    for j in range(n):
      mat[i][j] = np.exp((-1.0/ell) *scipy.linalg.norm(X[i, :] - Y[j, :]))
  return mat

def ornstein_uhlenbeck(x, y, ell):
  return np.exp((-1.0/ell) *scipy.linalg.norm(x-y))
```

```python
X, y = sklearn.datasets.make_moons(n_samples=100)
colors = [None]*len(y)
for i in range(len(y)):
  if(y[i]):
    colors[i] = 'b'
  else:
    colors[i] = 'r'
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.title("O conjunto de dados")
```

```python
std = np.std(X, axis=0) # Desvio padrão em cada componentte
ell = scipy.linalg.norm(std) # Fazemos ell = norma do vetor de desvios padrões
oh_gram_matrix = ornstein_uhlenbeck_gram(X, X, ell)
svm_oh_kernel = sklearn.svm.SVC(kernel='precomputed')
svm_rbf_kernel = sklearn.svm.SVC(kernel='rbf', gamma=gamma)
svm_oh_kernel.fit(oh_gram_matrix, y.T)
svm_rbf_kernel.fit(X, y.T)
```

## P3) Usando PCA para compressão de imagens

Considere uma imagem em preto-e-branco armazenada em uma matriz $\mathbf{X}$, de tamanho $n \times p$. Assuma que cada linha de $\mathbf{X}$ é uma obsevação em $\mathbb{R}^p$, fingindo então que $\mathbf{X}$ é uma matriz de dados observados. Usando a relação que vimos em sala de aula entre o PCA e a decomposição em valores singulares de uma matriz (SVD), mostre como podemos usar a SVD para comprimir uma imagem, com exemplos. Interprete esse procedimento do ponto de vista do PCA. Disserte amplamente sobre o observado.

A seguinte ideia é baseada nos links fornecidos acima e em $[1]$. 

Dado um conjunto de imagens, vamos tratar cada imagem como um vetor $\mathbf{x}_i$. 

Seja $\mathbf{X}$ a matriz dos dados observados, tal que a $i$-ésima linha de $\mathbf{X}$ corresponde ao vetor $\mathbf{x}_i$. 

Seja $\overline{\mathbf{x}}$ o vetor média de $\mathbf{X}$ tal que a $i$-ésima componente de $\overline{\mathbf{x}}$ é a média amostral da $i$-ésima coluna de $\overline{\mathbf{x}}$. 

Considere então a matriz dos dados centralizados $\tilde{\mathbf{X}} = \mathbf{X} - \overline{\mathbf{x}}$.(pequeno abuso de notação aqui)

Assim, a matrix de covariância destes dados é $\tilde{\mathbf{X}}^T \tilde{\mathbf{X}}$.

Usando a decomposição SVD para $\tilde{\mathbf{X}} = \mathbf{U} \mathbf{S} \mathbf{V}^T$, nos obtemos que 

$$  \tilde{\mathbf{X}}^T \tilde{\mathbf{X}} = (\mathbf{U} \mathbf{S} \mathbf{V}^T)^T (\mathbf{U} \mathbf{S} \mathbf{V}^T) = \mathbf{V} \mathbf{S}^2 \mathbf{V}^T $$

Assim, os autovetores necessários para realizar o PCA são dados pelas linhas de $\mathbf{V}$. Sejam $\mathbf{v}_1, \ldots, \mathbf{v}_n$ os autovetores. Então podemos fazer uma aproximação $M$-dimensional de cada vetor $\tilde{\mathbf{x}}_i \in \tilde{\mathbf{X}}$ por $[1]$

$$ \tilde{\mathbf{x}}_i \approx  \sum_{i=1}^{M} (\tilde{\mathbf{x}}_i^T \mathbf{v}_i) \mathbf{v}_i $$

Somando a média de volta, obtemos a aproximação $M$-dimensional para ${\mathbf{x}}_i \in {\mathbf{X}}$:

$$ {\mathbf{x}}_i \approx \overline{\mathbf{x}} + \sum_{i=1}^{M} (\tilde{\mathbf{x}}_i^T \mathbf{v}_i) \mathbf{v}_i $$

Portanto o algoritmo para compressão de imagens é o seguinte:
1. Calcular $\overline{\mathbf{x}}$ e armazenar seu valor.
2. Calcular $\tilde{\mathbf{X}} = \mathbf{X} - \overline{\mathbf{x}}$.
3. Fazer a decomposição SVD de $\tilde{\mathbf{X}} = \mathbf{U} \mathbf{S} \mathbf{V}^T$
4. Ordenar os autovetores em $\mathbf{V}$ segundo ordem decrescente dos seus autovetores correspondentes. 
5. Aproximar $\mathbf{x}_i$ por $\overline{\mathbf{x}} + \sum_{i=1}^{M} (\tilde{\mathbf{x}}_i^T \mathbf{v}_i) \mathbf{v}_i$.

Estamos considerando cada imagem como um vetor $\mathbb{R}^p$. Sob o ponto de visto do PCA, estamos supondo que estas imagens(vetores) estão contidos em um subespaço de dimensão $M < p$. Assim como o PCA, nosso algoritmo busca projetar os dados no espaço que maximiza a variância das projeções. No contexto da teoria da informação, isso quer dizer que escolhemos o espaço para o qual há menor perda de informação. 

Por exemplo, considere um conjunto de diferentes imagens do MNIST. Primeiro, é de se esperar que haverão regiões que não são preenchidas na maior parte das imagens, como por exemplo as bordas(os dígitos costumas estar no centro da imagem). Isso nos aponta, intuitivamente, que, de fato, nem todas as dimensões do espaço $\mathbb{R}^p$ carregam muita informação. Na grande maioria das imagens as bordas terão o mesmo valor(por exemplo preto em RGB). Assim já podemos concondar que as imagens realmente não cobrem todo o $\mathbb{R}^p$ e devem, portatno, estar contidas em um subespaço de dimensão menor.  Essa é a redundância mais aparente. Mas é claro que podem existir muito mais coordenadas "redundantes", reduzindo ainda mais o span do conjunto de dados no $\mathbb{R}^n$. A ideia do PCA em geral, assim como para a compressão de imagens, é de identificar quais são as direções realmente relevantes, ou melhor, de ranquear as direções segundo sua relevância e, daí, podemos considerar apenas uma quantidade arbitrária $M < p$ de informação, desconsiderando as $p - M$ que não parecem ter muita relevância. 

Para as imagens vamos considerar as imagens de dígitos manuscritos do conjunto de dados do MNIST.

Primeiro, vamos considerar um conjunto de imagens todas contendo o mesmo dígito, pois, como veremos, os autovetores obtidos são bem interpretáveis. 

Começamos portanto com um conjunto de 100  imagens diferentes do dígito oito.

```python
# Seu código aqui
# Carregando data dos dígitos
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
N = 100
single_digit = np.zeros((N, 28*28), float)
i = 0
while(i < 100):
  if(y_train[j] == 8):
    single_digit[i, :] = x_train[j].reshape(28*28)
    i += 1
  j += 1
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 4))
for i in range(3):
  ax[i].imshow(single_digit[i, :].reshape(28, 28), cmap='gray')
fig.suptitle("Exemplos de imagens deste conjunto")
plt.show()
```

```python
#Centralizando os dados
mean = np.mean(single_digit, axis=0)
for i in range(N):
  single_digit[i, :] = single_digit[i, :] - mean
```

```python
U, Sigma, V = np.linalg.svd(single_digit)
```

Podemos visualizar os autovetores como imagens. Abaixo, mostramos as imagens dos autovetores associados aos maiores autovalores, ou seja, os que dão principal components mais importantes.

```python
fig, ax = plt.subplots(1, 4, figsize=(16,4))
for i in range(4):
  ax[i].imshow(V[i, :].reshape(28, 28), cmap='gray')
  ax[i].set_title("%d-esimo autovetor" %(i+1,))
fig.suptitle("Primeiros autovetores da matriz de covariância")
plt.show()
```

Aqui, podemos ver como os primeiros autovetores já proveem uma base bem sólida para um dígito oito(lembrando que neste conjunto temos apenas imagens do dígito oito).

```python
def compressed_digit(x, mean, eigenvectors, M):
  soma = 0.0
  for i in range(M):
    soma = soma + (np.dot(x, eigenvectors[i])) * eigenvectors[i]
  return soma + mean
```

```python
fig, ax = plt.subplots(2, 4, figsize=(16,8))
eigenvectors = [V[i, :] for i in range(V.shape[0])]
m_values = [5, 10, 20, 50, 100, 200, 500, 728]
k = 0
for i in range(2):
  for j in range(4):
    compressed = compressed_digit(single_digit[0, :], mean, eigenvectors, m_values[k])
    ax[i][j].imshow(compressed.reshape(28, 28), cmap='gray')
    ax[i][j].set_title("Usando os %d primeiros components" %(m_values[k], ))
    if(k == 7):
      ax[i][j].set_title("Imagem original")
    k += 1
plt.tight_layout()
```

Assim, podemos ver que, de fato, uma parte considerável da informação armazenada em cada imagem não é tão relevante. Note que mesmo com apenas 5 componentes já é possível identificar o dígito oito na imagem. É claro que conforme aumentamos o número de componentes o dígito fica mais evidente e a resoulução da imagem fica melhor. Porém, se identificar o dígito fosse o único objetivo, poderíamos reduzir a quantidade de informação armazenada de 784 para 5 ou 10. Isso representa uma compressão, ou um ganho de memória, enorme. 

Essa drástica redução ocorreu, em parte, porque consideramos um conjunto de imagens do mesmo dígito. Agora vamos consider um conjunto de dados com dígitos distintos. Neste caso, os autovetores não tem imagens tão interpretáveis, mas é claro que  nosso algoritmo de compressão de imagens ainda é válido. 

```python
N = 100
many_digits = np.zeros((N, 28*28), float)
for i in range(N):
    many_digits[i, :] = x_train[i].reshape(28*28)
```

Aqui podemos ver que quando consideramos um conjunto mais "variado" de imagens, os autovetores não são interpretáveis. Note, porém, que não é apenas a interpretabilidade dos autovetores que muda. Retornando à interpretação do PCA, considerar um conjunto de dados mais "variado" significa que este conjunto de dados span uma região maior de $\mathbb{R}^p$, ou seja, o suposto subconjunto no qual estes dados estão contidos tem dimensão maior. Consequentemente, nós vamos precisar de uma base maior para este subconjunto, ou seja, cada autovetor é agora um vetor de uma base maior e, portanto, carrega menos informação. Logo, é de se esperar que necessitaremos de valores maiores de $M$ para obter boas aproximações das imagens.

```python
fig, ax = plt.subplots(2, 4, figsize=(16,8))
eigenvectors_many = [V_many[i, :] for i in range(V.shape[0])]
m_values = [5, 10, 20, 50, 100, 200, 500, 728]
l = np.random.randint(0, N)
k = 0
for i in range(2):
  for j in range(4):
    compressed = compressed_digit(many_digits[l, :], mean_many, eigenvectors_many, m_values[k])
    ax[i][j].imshow(compressed.reshape(28, 28), cmap='gray')
    ax[i][j].set_title("Usando os %d primeiros components" %(m_values[k], ))
    if(k == 7):
      ax[i][j].set_title("Imagem original")
    k += 1
plt.tight_layout()
```

Neste caso, note como nas três primeiras imagens(usando 5, 10, 20 componentes) o dígito não é evidente, em contraste com o resultado obtido no conjunto anterior. Note que há poquíssima diferença entre a imagens aproximadas com 100 ou 200 componentes e a imagem original. *distorção : $\sum_{n_{clusters}} (X_i - Centroid_n)^2  \text{ for } X_i \in cluster_n$

```python
distorcao = []
n_grupos = np.arange(1,11)
for i in n_grupos:
    kmean = KMeans( n_clusters=i, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0 )
    kmean.fit(sheet_norm)
    distorcao.append(kmean.inertia_)


plt.plot(n_grupos, distorcao, marker='o')
plt.xlabel('Numero de clusters')
plt.ylabel('Distorcao')
plt.show()
```

Nosso metodo do cotovelo também nao ajudou muito, não há nenhuma mudança brusca na inclinação da curva que possa nos ajudar a escolher o numero de clusters.

Outras metricas para verificação de clusters são : completeness score e v measure score , mas essa metricas apenas verificam verdade absolutas sobre clusters e não é interessante no nosso caso para comparação.
