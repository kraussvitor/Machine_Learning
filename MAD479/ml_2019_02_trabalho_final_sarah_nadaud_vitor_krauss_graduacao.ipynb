A seguir, investigamos as correlações entre as variáveis.

```python
correlation_matrix = songs_dataframe.drop(labels='decade', axis=1).corr()
correlated = np.zeros((correlation_matrix.shape[0], correlation_matrix.shape[1]), float)
correlated_indexes = []
threshhold = 0.6
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        if(i != j):
            if(abs(correlation_matrix.iloc[i, j]) > threshhold):
                correlated[i, j] = 1.0
                if(i < j):
                    correlated_indexes.append((i, j))
```

Podemos ver que alguns pares de váriáveis tem uma correlação alta(em valor absoluto). Devemos levar isto em conta na hora de escolher e treinar modelos. #### Feature Selection

Repare que a dimensão dos dados é consideravelmente grande, $90$, para ser exato. Um ganho computacional pode ser atingido se, de alguma forma, formos capazes de identificar um subconjunto de variáveis efetivamente relevantes para predições e outro das variáveis que não são. Com este intuito, vamos, agora, implementar alguns métodos para feature selection, isto é, buscaremos avaliar, de alguma forma, quais das $90$ variáveis regressoras são efetivamente significativas para predição da variável do ano de lançamento(ou década) e quais efetivamente não carregam muita informação. Em geral, estes métodos se baseiam em treinar modelos de classificação que também são capazes, em alguma maneira, de medir a importância de cada variável regressora. Por exemplo, sabemos que o Lasso e o Ridge, por causa de suas penalizações, resultariam(teoricamente) em coeficientes zero nas variáveis pouco importantes. Além deles, implementaremos alguns outros.

Atenção: para feature selection, é necessário separar os dados em conjunto de treino e de teste de antemão. Treinar modelos e fazer feature selection utilizando os dados de teste seria incorreto, como é explicado em $[1]$(mais especificamene na seção $7.10.2$).

Começamos pelo Ridge.

#### Ridge

Feature selection com o Ridge é bem imediato. Para as variáveis consideradas não importantes pelo Ridge, por conta da penalização, os seus respectivos coeficiente resultarão no valor zero. 

Deve se comentar que tratando-se de um problema de classificação, o mais correto seria utilizar regressão logística com penalização $\ell^2$. Porém, em primeiro lugar, este é um problema multiclasse. Neste caso, na verdade a regressão logística é feita one-vesus-all, ou seja, devem treinadas $C-1$ regressões logísticas, onde $C$ é o número de classes. Como o conjunto de dados é muito grande, isto representa um complexidade computacional enorme. Por isso, para reduzir a complexidade do problema, vamos considerar a estimação da década de lançamento como um problema de regressão e utilizar o Ridge.

```python
# Particionando conjuntos de treino e teste
songs_X_train = np.array(songs_dataframe.drop(labels=['label', 'decade'], axis=1).iloc[0:463715, :])
songs_decades_train = np.array(songs_dataframe['decade'])[0:463715]
songs_years_train = np.array(songs_dataframe['label'])[0:463715]
songs_X_test = np.array(songs_dataframe.drop(labels=['label', 'decade'], axis=1).iloc[463715:, :])
songs_decades_test =  np.array(songs_dataframe['decade'])[463715:]
songs_years_test =  np.array(songs_dataframe['label'])[463715:]
```

```python
# Normalizando os dados e separando as colunas de labels
train_scaler = StandardScaler()
songs_X_train = train_scaler.fit_transform(songs_X_train)
songs_X_test = train_scaler.transform(songs_X_test) # Note que o conjunto de teste deve seguir a mesma normalização do conjunto de treino
```

```python
alphas = np.logspace(-3, 1, 50)
mean_square = make_scorer(mean_squared_error)
ridge = RidgeCV(alphas=alphas, scoring=mean_square)
start = time.time()
ridge.fit(songs_X_train, songs_years_train)
end = time.time()
```

```python
print("Tempo de execução do Ridge: %1.3f segundos" %(end-start))
print("Número de parametros zero: ", len(ridge.coef_) - np.count_nonzero(ridge.coef_))
```

Podemos ver que nenhum dos coeficientes estimados pelo Ridge foi igual a zero. Ou seja, pelo Ridge, não há nenhuma variável regressora que não seja significativa para predição dos valores do ano de lançamento e, portanto, não devemos arbitrariamente desconsiderar uma delas.

#### Lasso

Feature selection com o Lasso é similar ao Ridge, a única diferença sendo a penalização $\ell^1$. Novamente, vamos considerar como um problema de regressão por conta de sua menor complexidade computacional. 

```python
lasso = LassoCV(alphas=alphas)
start = time.time()
lasso.fit(songs_X_train, songs_years_train)
end = time.time()
```

```python
print("Tempo de execução do Lasso: %1.3f segundos" %(end-start))
print("Número de parametros zero: ", len(lasso.coef_) - np.count_nonzero(lasso.coef_))
print("Indice dos coeficientes iguais a zero: ", np.where(lasso.coef_==0)[0])
```

Podemos ver que com o Lasso seis coeficientes foram iguais a zero.

Assim, vemos que tanto o Ridge quanto o Lasso consideraram a grande maioria das variáveis regressoras como relevantes. De fato, o Lasso nos proporcionaria uma redução de apenas $6$ das $90$ dimensões. Isso não nos daria um ganho computacional muito significativo. 

Com efeito, o Ridge e o Lasso foram apenas a linha de frente da nossa feature selection. Diante da discordância destes dois métodos na quantidade de regressoras relevantes, além da redução não significativa que eles nos proporcionariam, vamos estender nossa análise com mais métodos. Começamos com Random Forests.

*Obs:* o Ridge e o Lasso involvem o uso de um parâmetro(learning rate) associado a penalização. O treinamento de um único Lasso ou Ridge é bem pequeno, mas aqui incluímos nos seus tempos de execução todo o processo de otimização do parâmetro por cross-validation.

#### Random Forest Classifier

Lembre que para cada nó em uma Decision Tree $[1]$ está associado um valor chamado 'impurity', que geralmente tomamos como sendo Gini-impurity ou Information-gain. No caso da Gini-impurity, por exemplo, este valor mede o quão heterogêneo, em relação às classes, o subconjunto(multidimnesional) definido por aquele nó é. Um subconjunto seria puro(homogêneo) se todos os dados contidos nele fossem da mesma classe. Sem entrar no mérito de overfitting, a melhor Decision Tree(pelo menos para o conjunto de treino) teria todos os seus subconjuntos puros(se não restringíssemos sua altura). Em cada nó fazemos um split em uma das variáveis regressoras que melhor separa os dados segundo suas classes, no caso o que mais reduz a impuridade. Depois que todas as Decision Trees foram treinadas, podemos, para cada split em cada Tree, considerar a redução na impurity causada por este split e acumulamos estes valores respectivamente à feature escolhida. Isso nós dá uma medida da importância de cada feature.(mais detalhes nos capítulos de GradientBoosting e RandomForests em $[1]$). 

```python
RandomForest = RandomForestClassifier(n_estimators=10)
start = time.time()
RandomForest.fit(songs_X_train, songs_decades_train)
end = time.time()
```

```python
forest_importances = pd.Series(RandomForest.feature_importances_, index=songs_dataframe.drop(labels=['label', 'decade'], axis=1).columns)
forest_importances.nlargest(20).plot(kind='barh')
plt.title("20 componentes mais importantes")
plt.show()
```

```python
print("Tempo de execução do Random Forest: %1.3f segundos" %(end-start))
forest_importances.plot(kind='barh')
frame = plt.gca()
#frame.axes.get_xaxis().set_visible(False)
frame.axes.get_yaxis().set_visible(False)
plt.title("Importância de cada regressora")
plt.show()
```

#### Particle Swarm Optimization

Algoritmos genéticos e evolucionários também já foram propostos para feature selection $[7]$. Nós vamos implementar o proposto pelo em $[2]$. A ideia é otimizar um vetor binário $x = (x_1, \ldots, x_{90})$ no qual a componente $x_i$ denota se a variável(feature) deve ser considerada($x_i=1$) ou não($x_i=0$). Para a otimização, vamos utilizar Particle Swarm Optimization. Assim como proposto em $[2]$, consideramos a seguinte função de custo:

$$ f(x) = \alpha (1-p) + (1 - \alpha)\Big( 1 - \frac{N_f}{N_t} \Big) $$

onde $\alpha$ é um hiperparâmetro que mede o tradeoff entre a performance $p$ do classificador e o número de features $N_f$. Além disso, $N_t$ é o número total de features. Neste caso, $N_t=90$.

```python
# Considerando apenas um subconjunto dos dados
X = songs_X_train[0:1000, :]
y = songs_decades_train[0:1000]
```

```python
def particle_loss_function(binary, X, y, clf, alpha):
    if np.count_nonzero(binary) == 0:
        subset_X = np.array(X)
    else:
        subset_X = np.array(X)[:,binary==1]
    clf.fit(subset_X, np.array(y))
    p = (clf.predict(subset_X) == y).mean()
    return (alpha * (1.0 - p) + (1.0 - alpha) * (1 - (subset_X.shape[1] / len(binary))))
```

```python
def swarm_loss_function(X, y, clf, alpha=0.5):
    def loss_function(binaries):
        return np.array([particle_loss_function(binaries[i], X, y, clf, alpha) for i in range(len(binaries))])
    return loss_function
```

```python
classifier = linear_model.LogisticRegressionCV()
options = {'c1': 0.5, 'c2': 0.5, 'w':0.9, 'k': 5, 'p':2}
optimizer = pyswarms.discrete.BinaryPSO(n_particles=10, dimensions=90, options=options)
function = swarm_loss_function(X, y, classifier, alpha=0.75)
start = time.time()
optimal_cost, activations = optimizer.optimize(function, iters=10)
end = time.time()
```

É importante ressaltar que o parâmetro $\alpha$ vai alterar o número de variáveis selecionadas, como pode ser visto da função de custo considerada. Aqui, podemos ver que para $\alpha = 0.75$ este número foi $62$, ou seja, de acordo com este método, $62$ das $90$ variáveis regressoras foram consideradas relevantes. É claro que um melhor ajuste do parâmetro $\alpha$ pode ser obtido por cross-validation, sob um pesado custo computacional de realizar uma tarefa de otimização diversas vezes.

#### Isomap

Diferentemente do PCA, que faz uma projeção linear dos dados(projeta os dados em um hiperplano), o Isomap e o LLE fazem transformações não lineares dos dados. Ambos estes métodos buscam projetar os dados em uma variedade, em geral não linear, contida em um subespaço de dimensão menor que o espaço original.

De forma breve, o Isomap é similar ao Multidimensional Scaling $[1]$, a principal diferença sendo que o Isomap utiliza a distância geodésica. Como a suposta variedade na qual os dados estão contidas é desconhecida, o Isomap estima as distâncias geodésicas por meio de um grafo. Neste grafo, cada um dos dados é adjacente a seus $k$ vizinhos mais próximos, e o peso associado a cada aresta é igual a distância euclidiana entre os dois dados. Com isso, as distâncias geodésicas entre dois pontos é estimada pelo tamanho(soma dos pesos das arestas) do menor caminho no grafo entre estes dois pontos(mais detalhes sobre o Isomap em $[1,3]$). Sendo assim, a escolha do parâmetro $k$ tem grande influência na projeção obtida pelo Isomap. Ele ajusta quão local o método será. Note que se $k$ é igual ao número de dados a distância geodésica fica igual à euclidiana e o método se torna global. Métodos para escolha de $k$ são propostos em $[3,4]$. Aqui, nós implementamos o proposto em $[3]$. 

```python
def optimal_isomap_parameter(k_low, k_up, data, num_components, solver='dense'):
    k_range = range(k_low+1, k_up+1)
    isomap_zero = Isomap(n_neighbors=k_low, n_components=num_components, eigen_solver=solver)
    isomap_one = Isomap(n_neighbors=k_range[1], n_components=num_components, eigen_solver=solver)
    isomap_zero.fit_transform(data)
    isomap_one.fit_transform(data)
    f_k_minus_one = np.sum(np.sum(isomap_zero.dist_matrix_))
    f_k = np.sum(np.sum(isomap_one.dist_matrix_))
    for i in range(2, len(k_range)):
        isomap = Isomap(n_neighbors=k_range[i], n_components=num_components, eigen_solver=solver)
        isomap.fit_transform(data)
        temp = np.sum(np.sum(isomap.dist_matrix_))
        if(f_k - temp > 2*(f_k_minus_one - f_k)):
            return k_range[i]
        f_k_minus_one = f_k
        f_k = temp
    return k_up
```

#### Locally Linear Embedding

Similarmente ao Isomap, o LLE tem o mesmo parâmetro $k$ que determina o número de vizinhos mais próximos que serão adjacentes a cada dado no grafo associado. A diferença entre os dois métodos é que no LLE cada ponto é aproximado, ou reconstruído, utilizando estes $k$ vizinhos. Primeiro, o LLE calcula o conjunto ótimo de pesos $\omega$ que fornece a melhor reconstrução de cada dado. Depois disso, o LLE prossegue de forma similar ao Isomap, buscando os vetores(projetados) que melhor mantém a geometria local dos $k+1$ vizinhos no espaço original(mais detalhes sobre o LLE em $[1,5]$ e especialmente em $[6]$). Novamente, a escolha do parâmetro $k$ terá efeito na qualidade da projeção dada pelo LLE. Em $[5]$ é proposto um algoritmo para escolha de $k$, que implementamos abaixo.  

```python
def optimal_lle_parameter(k_low, k_up, data, num_components, solver='dense'):
    k_range = range(k_low, k_up+1)
    residual_vars = np.zeros((len(k_range)))
    for i in range(len(k_range)):
        lle = LocallyLinearEmbedding(n_neighbors=k_range[i], n_components=num_components, eigen_solver=solver)
        output = lle.fit_transform(data)
        D_x = euclidean_distances(data)
        D_y = euclidean_distances(output)
        diff = D_x - D_y
        residual_vars[i] = np.var(np.array([diff[i, :] for i in range(diff.shape[0])]))/np.var(np.array([D_x[i, :] for i in range(D_x.shape[0])]))
    return k_range[np.argmin(residual_vars)]
```

```python
lle_opt_k = optimal_lle_parameter(k_low=2, k_up=30, data=subset_X_train[selected_indices, :], num_components=20)
print("Valor ótimo de k para o LLE: ", lle_opt_k)
```

```python
lle = LocallyLinearEmbedding(n_neighbors=lle_opt_k, n_components=20, eigen_solver='dense')
start = time.time()
lle_X = lle.fit_transform(subset_X_train)
end = time.time()
print("Tempo de execução do LLE: %1.4f segundos" %(end-start))
```

#### Autoencoder

Como último método de redução de dimensionalidade vamos utilizar autoencoders. Autoencoders são originalmente usados para compressão de dados como imagens ou áudio, mas a ideia é similar aos demais métodos: projetar, de alguma forma, os dados em um subespaço de dimensão menor, mas buscando de fazer isso de forma a perder a menor quantidade de informação. No caso, nossos autoencoders serão feitos por meio de redes neurais feedforward. Isto é, vamos construir uma rede neural que recebe como entrada os nossos dados, no espaço $\mathbb{R}^{83}$, e terá como saída um vetor em $\mathbb{R}^{20}$. Esta rede é chamada 'encoder'. Teremos também uma segunda rede neural, chamada 'decoder', que recebe como entrada um vetor em $\mathbb{R}^{20}$ e tem como saída um vetor no espaço original $\mathbb{R}^{83}$. As redes neurais são então treinadas conjuntamente, buscando minimizar o erro quadrático médio entre o vetor decodificado e o original(na verdade, usando o keras para implementação, vamos treinar uma única rede neural que é a "composição" das duas). 

```python
def init_autoencoder(p=83, d=20):
    #encoding part
    autoencoder = Sequential()
    autoencoder.add(Dense(units=p, activation='sigmoid', use_bias=True, input_shape=(p,)))
    autoencoder.add(Dense(units=int(p/2), activation='tanh', use_bias=True))
    autoencoder.add(Dense(units=d, activation='tanh', use_bias=True))
    #decoding part
    autoencoder.add(Dense(units=int(p/2), activation='tanh', use_bias=True))
    autoencoder.add(Dense(units=p, activation='sigmoid', use_bias=True))
    return autoencoder
```

```python
autoencoder = init_autoencoder(83, 20)
autoencoder.compile(optimizer='adamax', loss='mse')
```

```python
start = time.time()
training_hist = autoencoder.fit(x=subset_X_train, y=subset_X_train, epochs=350, batch_size=200, verbose=0)
end = time.time()
print("Tempo de execução do autoencoder: %1.4f segundos" %(end-start))
```

Começamos com uma análise da complexidade computacional de cada método. Como já era se esperar, o PCA apresentou um tempo de execução bem menor que os demais. De fato a complexidade do PCA é $\mathcal{O}(N^2 p + p^3) = \mathcal{O}(N^2 p)$, onde $N$ é o tamanho do conjunto e $p$ a dimensão dos dados. O Isomap e o LLE involvem encontrar caminhos mais curtos em um grafo com $N$ nós e $Nk$ arestas, o que tem complexidade $\mathcal{O}(N^2 k^2 + N \log N)$ com o algoritmo de Dijkstra. A segunda etapa do algoritmo involve otimizar as $N$ projeções com dimensão $d$, o que tem complexidade $\mathcal{O}(N^2d)$. Finalmente, para uma rede neural com $\ell_k$ neurônios na $k$-ésima camada e $e$ epochs, a complexidade de treino da rede é $\mathcal{O}(N \;e \; \sum \ell_{k} \ell_{k+1})$.

Assim, concluímos que os tempos de execução obtidos respeitaram a ordem de complexidade teórica dos métodos. Em termos de complexidade, o autoencoder se demonstrou, portanto, uma boa alternativa ao Isomap e LLE. Lembramos aqui que tivemos que considerar apenas um subconjunto dos dados por conta justamente do tempo de execução destes dois métodos em nossos computadores pessoais. Contudo, mais importante que a complexidade de cada um dos métodos é o quão bem eles foram capazes de reduzir a dimensionalidade dos dados com a menor perda de informação. Dito isso, prosseguimos então com a próxima parte do nosso trabalho, que é de treinar e avaliar modelos utilizando estes novos dados projetados.

#### Neural Networks

Novamente, escolhemos um método capaz de fornecer fronterias de decisão não-lineares. A escolha por redes neurais se dá, principalmente, por conta de sua complexidade computacional relativamente menor, se comparada a outros métodos de classificação não lineares. Como o objetivo final é treinar um modelo no conjunto de treino inteiro, este é um fator muito relevante. O treinamento de uma SVM, por exemplo, tem complexidade quadrática no número de dados. Para o caso deste conjunto, com quase meio milhão de dados de treino, isto se torna inviável em computadores pessoais. Por isso, esperamos que as redes neurais se provem como alternativas confiáveis para este volume tão grande de dados.

```python
def init_neural_net(input_size, output_size):
    neural_net = Sequential()
    neural_net.add(Dense(units=int(2*input_size), activation='relu', use_bias=True, input_shape=(input_size,)))
    neural_net.add(Dense(units=int(input_size), activation='relu', use_bias=True))
    neural_net.add(Dense(units=int(0.5*input_size), activation='relu', use_bias=True))
    neural_net.add(Dense(units=int(output_size), activation='softmax', use_bias=True))
    return neural_net
```

- Tanto o LLE quanto o Isomap buscam aproximar distâncias geodésicas por meio de distâncias em um grafo. Se o dados são densos no espaço, esta parece ser uma boa aproximação. Se os dados são densos, a semireta que liga dois pontos vai estar aproximadamente contida na variedade e, portanto, a distâncias no grafo serão boas aproximações das geodésicas. Porém, se os dados são esparsos, a semireta ligando dois pontos não vai estar contida na variedade("ela vai cortar caminho") e, assim, obteremos estimativas ruins. Por isso, estes métodos parecem necessitar de uma quantidade maior de dados, para garantir esta densidade, ou alguma indicação prévia de que os dados estão de fato próximos de uma variedade. 
- Além disso, o LLE e o Isomap requerem a escolha do parâmetro $k$ de vizinhos. Há algumas propostas de algoritmos para encontrar este valor. Porém, devido a alta complexidade destes métodos, a otimização deste parâmetro tem um custo computacional enorme. Aqui, tivemos que estimar o parâmetro com um número ainda mais reduzido de dados, o que pode ter comprometido a escolha de um bom valor.

Uma possível forma de tentar lidar com este problema é fazer amostragem sintética, isto é, para uma classe com menor número de dados, nós geramos novos dados sintéticamente e os atribuímos à esta classe. É evidente que este dados devem ser gerados de uma maneira tal que eles sejam, de alguma forma, similares aos verdadeiros dados da classe. Uma maneira de fazer isso é utilizando nearest neighbours, método que é conhecido como SMOTE.
